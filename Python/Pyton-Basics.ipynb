## Expression, Statement, dan Variabel

```python
# Ekspresi adalah kombinasi nilai dan operator untuk menghasilkan suatu nilai
# (sering disebut operasi)

"""
Operator Aritmatika:
+   = Penjumlahan
-   = Pengurangan
*   = Perkalian
/   = Pembagian
//  = Pembagian bulat
**  = Pangkat
%   = Modulus (sisa bagi)

Operator Logika:
and = Logika AND
or  = Logika OR
not = Logika NOT

Operator Bitwise:
& = Bitwise AND
| = Bitwise OR

Operator Perbandingan:
<   = Kurang dari
>   = Lebih dari
<=  = Kurang dari atau sama dengan
>=  = Lebih dari atau sama dengan
==  = Sama dengan
!=  = Tidak sama dengan
"""

# Statement adalah perintah yang dijalankan oleh program
print(f"Hasil dari 3 ** 4 adalah {3 ** 4}")

# Variable adalah tempat untuk menyimpan data
x = 3
y = 4 // 2
print(f"Hasil dari {x} + {y}  adalah {x + y}")

# Aturan Matematika tetap berlaku
print(f"Hasil dari {x} + {y} ** {x} adalah {x + y ** x} dan 7 + (4 + 3) adalah {7 + (4 + 3)}")
```

## Type Data

```python
# String (str) → digunakan untuk menyimpan teks
"""
apapun isinya kalo diapit dengan tanda petik atau ada spasi maka tetap string, misal :
"%^$#%" dan "1 2 3 4"
"""

kata = "Hello Python"
print("Nilai kata       :", kata)
print("Tipe data kata   :", type(kata))
print()

# Integer (int) → digunakan untuk menyimpan bilangan bulat
angka = 10
print("Nilai angka      :", angka)
print("Tipe data angka  :", type(angka))
print()

# Float (float) → digunakan untuk menyimpan bilangan desimal
angka_desimal = 3.14
print("Nilai desimal    :", angka_desimal)
print("Tipe data desimal:", type(angka_desimal))
print()

# Boolean (bool) → hanya memiliki dua nilai: True atau False
salah = False
benar = True
print("Nilai boolean    :", benar)
print("Tipe data boolean:", type(benar))
print("Nilai boolean    :", salah)
print("Tipe data boolean:", type(salah))
print()


# KONVERSI TIPE DATA
print("Konversi float ke int   :", int(angka_desimal))   # 3
print("Konversi bool ke string :", str(benar))           # "True"
print("Konversi int ke float   :", float(angka))         # 10.0
print("Konversi string ke bool :", bool(kata))           # True (karena tidak kosong)

"""
Konversi tipe data hanya dapat dilakukan jika nilainya masuk akal (valid).

Contoh:
- "123"  → bisa dikonversi menjadi int
- "1"    → bisa dikonversi menjadi int
- "Saya" → TIDAK bisa dikonversi menjadi int karena bukan angka

Jika dipaksakan, program akan menghasilkan error (ValueError).
"""

print()
```

## Indexing

```python
# Mengambil 1 item
Nama = "Affan Suhendar"
print(f"Index ke 4 dari {Nama} adalah {Nama[4]}")

# Mengambil index dari belakang
print(f"Index ke -1 dari {Nama} adalah {Nama[-1]}")

# Mengambil rentang index
print(f"Index ke 2 sampai 5 dari {Nama} adalah {Nama[0:5]}")

# Index dengan lompatan
print(f"{Nama[::2]}")

# Index dengan lompatan negatif
print(f"{Nama[::-2]}")
```

## String Operation

```python
# Nambah
Nama_Depan = "Affan"
Nama_Belakang = "Suhendar"
Nama_Lengkap = Nama_Depan + " Suhendar"
print(Nama_Lengkap)
```

```python
# Upper
print(f"Hasil dari Nama_Lengkap.upper() adalah '{Nama_Lengkap.upper()}'")

# Lower
print(f"Hasil dari Nama_Lengkap.lower() adalah '{Nama_Lengkap.lower()}'")

# Capitalize
print(f"Hasil dari Nama_Lengkap.capitalize() adalah '{Nama_Lengkap.capitalize()}'")

# Replace
print(f"Hasil dari Nama_Lengkap.replace('Suhendar', 'Kurniawan') adalah '{Nama_Lengkap.replace('Suhendar', 'Kurniawan')}'")

# find
print(f"Hasil dari Nama_Lengkap.find('Suhendar') adalah {Nama_Lengkap.find('Suhendar')}")
```

## String Formatting

```python
print("Berat Badan saya adalah {} kg".format(75))

# F-String
print(f"Berat Badan saya adalah {75} kg")

# Format
print("Berat Badan saya adalah {Berat} kg".format(Berat=75))
```

## Operator Perbandingan + If Else

```python
"""
<   = Kurang dari
>   = Lebih dari
<=  = Kurang dari atau sama dengan
>=  = Lebih dari atau sama dengan
==  = Sama dengan
!=  = Tidak sama dengan
"""

# Perbandingan
print(f"Apakah 1 == 2 ? {1 == 2}")
print(f"Apakah 1 != 2 ? {1 != 2}")
print(f"Apakah 1 > 2 ? {1 > 2}\n")

# Case Nilai
Affan = 90

if Affan >= 87:
    print("Nilai kamu A") 

elif Affan >= 77:
    print("Nilai kamu B")

elif Affan >= 67:
    print("Nilai kamu C")

elif Affan >= 57:
    print("Nilai kamu D")

else:
    print("Nilai kamu E")

```

```python
# Contoh 1 line
nilai = 8.5
print("nilai tertinggi" if nilai > 9 else "Nilai kamu rendah")

nilai = 5
print("Nilai aku ", "Tinggi" if nilai > 9 else "Rendah")
```

## Enumerate

```python
# Enumerate Menghasilkan index dan item dalam list
tahun = [2000, 2001, 2002, 2003]
for i, item in enumerate(tahun):
    print(f"Index ke-{i} adalah {item}")
```

```python
# Contoh lain dari enumerate : Mindahin Data 
box = []
Gudang = [3,4,5,6]

for i, j in enumerate(Gudang):
    box.append(j)

print(box)
```

## While Loop

```python
i  = 0 

while i < 5 :
    print(i)
    i += 1
```

# For Else

```python
# Else akan dieksekusi kalo tidak ada break didalam loop
for i in range(3):
    print(i)
else:
    print("Selesai\n\n")

# Break akan dieksekusi kalo ada break didalam loop
for i in range(3):
    print(i)
    break
else:
    print("Selesai")
```

## Nested Loop

```python
Kotak = [[1,2,3], ["Aku","Suka","Kamu"]]

for i in Kotak:
    for j in i:
        print(j)
    print()
```
