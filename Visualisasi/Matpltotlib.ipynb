## Matplotlib sederhana

```python
import matplotlib.pyplot as plt

# Plot (X,Y)
plt.plot([1,2,3,4], [1,4,9,16])
plt.xlabel("Nilai X")
plt.ylabel("Nilai Y")
plt.title("Plot Sederhana")
plt.show()
```

## Matplotlib Versi OOP

```python
import matplotlib.pyplot as plt

fig = plt.figure()
axes = fig.add_axes([0.1,0.3,0.5,0.5]) # left, bottom, width, height
"""
0.1 = 10% 
"""

axes.plot([1,2,3,4], [1,4,9,16])
axes.set_xlabel("Nilai x")
axes.set_ylabel("Nilai y")
axes.set_title("Plot Sederhana")
plt.show()
```

## Versi 1

```python

# Bikin kanvas
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])

# Plot axes 1
axes1.plot([1,2,3,4], [1,4,9,16], color="red")
axes1.set_xlabel("Nilai x")
axes1.set_ylabel("Nilai y")
axes1.set_title("Plot Sederhana")

# Plot axes 2
axes2.plot([1,2,3,4], [16,9,4,1])
axes2.set_xlabel("Nilai x")
axes2.set_ylabel("Nilai y")
axes2.set_title("Plot Sederhana")

plt.show()
```
